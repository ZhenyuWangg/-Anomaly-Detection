> Create a function to visulize variable importance

```python
def VarImp(model_name):
    
    from sklearn.metrics import roc_curve,auc
    from sklearn.metrics import average_precision_score
    from sklearn.metrics import precision_recall_curve
    import matplotlib.pyplot as plt
    
    # plot the variable importance
    plt.rcdefaults()
    variables = model_name._model_json['output']['variable_importances']['variable']
    y_pos = np.arange(len(variables))
    fig, ax = plt.subplots(figsize = (6,len(variables)/2))
    scaled_importance = model_name._model_json['output']['variable_importances']['scaled_importance']
    ax.barh(y_pos,scaled_importance,align='center',color='green')
    ax.set_yticks(y_pos)
    ax.set_yticklabels(variables)
    ax.invert_yaxis()
    ax.set_xlabel('Scaled Importance')
    ax.set_title('Variable Importance')
    plt.show()
```

> Define a `createGains` Function, for repetition purpose of generate the gain table 

```python
def createGains(model):
    predictions = model.predict(test_hex)
    test_scores = test_hex['loan_default'].cbind(predictions).as_data_frame()

    #sort on prediction (descending), add id, and decile for groups containing 1/10 of datapoints
    test_scores = test_scores.sort_values(by='predict',ascending=False)
    test_scores['row_id'] = range(0,0+len(test_scores))
    test_scores['decile'] = ( test_scores['row_id'] / (len(test_scores)/10) ).astype(int)
    #see count by decile
    test_scores.loc[test_scores['decile'] == 10]=9
    test_scores['decile'].value_counts()

    #create gains table
    gains = test_scores.groupby('decile')['loan_default'].agg(['count','sum'])
    gains.columns = ['count','actual']
    gains

    #add features to gains table
    gains['non_actual'] = gains['count'] - gains['actual']
    gains['cum_count'] = gains['count'].cumsum()
    gains['cum_actual'] = gains['actual'].cumsum()
    gains['cum_non_actual'] = gains['non_actual'].cumsum()
    gains['percent_cum_actual'] = (gains['cum_actual'] / np.max(gains['cum_actual'])).round(2)
    gains['percent_cum_non_actual'] = (gains['cum_non_actual'] / np.max(gains['cum_non_actual'])).round(2)
    gains['if_random'] = np.max(gains['cum_actual']) /10 
    gains['if_random'] = gains['if_random'].cumsum()
    gains['lift'] = (gains['cum_actual'] / gains['if_random']).round(2)
    gains['K_S'] = np.abs( gains['percent_cum_actual'] -  gains['percent_cum_non_actual'] ) * 100
    gains['gain']=(gains['cum_actual']/gains['cum_count']*100).round(2)
    gains = pd.DataFrame(gains)
    return(gains)
```

```python
createGains(glm_v1)
```

> Define the `ROC_AUC` Function to plot the Roc Curve

```python
def ROC_AUC(my_result,df,target):
    from sklearn.metrics import roc_curve,auc
    from sklearn.metrics import average_precision_score
    from sklearn.metrics import precision_recall_curve
    import matplotlib.pyplot as plt

    # ROC
    y_actual = df[target].as_data_frame()
    y_pred = my_result.predict(df).as_data_frame()
    fpr = list()
    tpr = list()
    roc_auc = list()
    fpr,tpr,_ = roc_curve(y_actual,y_pred)
    roc_auc = auc(fpr,tpr)
    
    # Precision-Recall
    average_precision = average_precision_score(y_actual,y_pred)

    print('')
    print('   * ROC curve: The ROC curve plots the true positive rate vs. the false rositive sate')
    print('')
    print('   * The area under the curve (AUC): A value between 0.5 (random) and 1.0 (perfect), measuring the prediction accuracy')
    print('')
    print('   * Recall (R) = The number of true positives / (the number of true positives + the number of false negatives)')
    print('')
    
    # plotting
    plt.figure(figsize=(10,4))

    # ROC
    plt.subplot(1,2,1)
    plt.plot(fpr,tpr,color='darkorange',lw=2,label='ROC curve (aare=%0.2f)' % roc_auc)
    plt.plot([0,1],[0,1],color='navy',lw=3,linestyle='--')
    plt.xlim([0.0,1.0])
    plt.ylim([0.0,1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic: AUC={0:0.4f}'.format(roc_auc))
    plt.legend(loc='lower right')

    # Precision-Recall
    plt.subplot(1,2,2)
    precision,recall,_ = precision_recall_curve(y_actual,y_pred)
    plt.step(recall,precision,color='b',alpha=0.2,where='post')
    plt.fill_between(recall,precision,step='post',alpha=0.2,color='b')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.ylim([0.0,1.05])
    plt.xlim([0.0,1.0])
    plt.title('Precision-Recall curve: PR={0:0.4f}'.format(average_precision))
    plt.show()
```

```python
ROC_AUC(glm_v1,test_hex,'loan_default')
```

> - Find the best Hyperparameter using grid search <a class="anchor" id="section-3-5"></a>

```python
from h2o.grid.grid_search import H2OGridSearch

# Define hyperparameters for grid search
hyper_params = {
    'alpha': [0.1, 0.3, 0.5],  # Regularization parameter
    'lambda': [1e-6, 1e-9, 1e-12]  # Regularization strength
}

# Create a dictionary of parameters
search_criteria = {'strategy': 'Cartesian'}

# Define the grid search
glm_grid = H2OGridSearch(model=H2OGeneralizedLinearEstimator(model_id='glm_v2', seed=1234),
                          hyper_params=hyper_params,
                          search_criteria=search_criteria)

# Train the grid search
glm_grid.train(x=predictors, y=target, training_frame=train_hex)

# Get the best model from the grid search
glm_v2 = glm_grid.get_grid()[0]
glm_v2
```

## Section 5: Model Summary & Conclusion <a class="anchor" id="section-5"></a>

### Model `GLM`
- Gradient Boosting Machines (GBM) play a pivotal role in mortgage default prediction and anomaly detection. GBM's ensemble learning sequentially corrects errors of previous models, fostering a robust predictive model for default probabilities in mortgage applications. Its iterative nature identifies high-risk applications, aiding lenders in proactive risk assessment by flagging instances where predicted default probabilities deviate significantly from historical patterns. This anomaly detection capability prompts investigations into potential outliers, allowing financial institutions to apply targeted risk mitigation strategies. By combining GBM's predictive strength with anomaly detection, lenders gain insights into both anticipated defaults and irregularities, facilitating proactive risk management strategies and maintaining a healthier mortgage portfolio.

- GLM's anomaly detection capabilities highlight outliers by identifying instances where predicted default probabilities significantly diverge from expected norms. The model's transparency aids in understanding these anomalies, revealing the influential predictors behind these deviations and enabling further investigation into atypical cases, ensuring proactive risk management within mortgage portfolios.

### Model `AutoML`
- AutoML, an automated machine learning framework, revolutionizes the mortgage industry's risk assessment by efficiently handling model selection, hyperparameter tuning, and anomaly detection. In mortgage default prediction, AutoML conducts an exhaustive exploration across diverse algorithms such as Gradient Boosting Machines (GBM), Generalized Linear Models (GLM), Random Forests, and Neural Networks. By leveraging this array of models, it identifies the most accurate predictor of default probabilities, ensuring lenders have access to the most effective predictive tools for risk assessment. Simultaneously, AutoML's versatility extends to anomaly detection, meticulously examining discrepancies between predicted and expected outcomes across multiple models. This process allows for the identification of mortgage applications where predictions deviate significantly from historical norms, signaling potential anomalies or outliers requiring closer scrutiny. AutoML's holistic approach amalgamates various modeling techniques, empowering institutions to proactively manage risks by pinpointing atypical cases that might pose threats to mortgage portfolios.
